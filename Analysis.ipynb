```python
import seaborn as snb


sns_data = train[train['seat']==0].drop(['time', 'seat', 'event'], axis=1)
correlations = sns_data.corr()
snb.heatmap(correlations, linewidths=.5, annot=True)

plt.gcf().set_size_inches(16, 12)
plt.title("Heatmap: Seat 0", size=20)
plt.show()

sns_data = train[train['seat']==1].drop(['time', 'seat', 'event'], axis=1)
correlations = sns_data.corr()
snb.heatmap(correlations, linewidths=.5, annot=True)

plt.gcf().set_size_inches(16, 12)
plt.title("Heatmap: Seat 1", size=20)
plt.show()
```

```python
train = train_one_pair
train = train.drop(['crew'], axis=1)

data_first_pilot = train[train['seat'] == 1]
data_second_pilot = train[train['seat'] == 0]

f_train = data_first_pilot.drop('seat', axis=1)
s_train = data_second_pilot.drop('seat', axis = 1)

print("Shape of data: \n first pilot -  ", f_train.shape, "\n second pilot - ", s_train.shape)
```
